```python


n_student = 100
study_hour = np.random.uniform(low = 0, high = 9, size = (n_student,))
math_score = 10 * study_hour + 10

plt.xlabel('study_hour')
plt.ylabel('math_score')
plt.title('study_hour - math_score relation')
plt.grid(True)
plt.xlim(0,20)
plt.ylim(0,200)
plt.scatter(study_hour, math_score)

```
